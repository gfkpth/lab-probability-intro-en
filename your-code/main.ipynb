## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
n_white = 10
n_red = 12
n_black = 8

n_total = n_white + n_red + n_black

#
p_white = n_white/n_total
print(f'The probability of taking a white ball out is: {p_white}')

# dependent probability without replacement
# probability of extracting white ball on first draw is p_white
# probability of drawing black ball from total set decremented by 1 is p_black_minus_one
# dependent probability is product of p_white and p_black_minus_1 divided by probability of preceding event (p_white)
p_black_minus1 = n_black/(n_total - 1)
p_black_after_white_noreplace = p_white * p_black_minus1
print(f'The probability of taking a black ball out after a first one without replacing: {p_black_after_white_noreplace}')


# independent cumulative probability
# product of probability of drawing white and drawing black from identical total
p_black = n_black/n_total
p_black_and_white_replace = p_black * p_white
print(f'The probability of drawing a white and a black ball with replacing: {p_black_and_white_replace}')

#
p_red = n_red/n_total

# Consider the possible ways of getting three different colours

p_each_color_once_v1 = p_red * ( (n_white/(n_total-1))*(n_black/(n_total-2)) + (n_black/(n_total-1))*(n_white/(n_total-2)))
p_each_color_once_v2 = p_black * ( (n_white/(n_total-1))*(n_red/(n_total-2)) + (n_red/(n_total-1))*(n_white/(n_total-2)))
p_each_color_once_v3 = p_white * ( (n_red/(n_total-1))*(n_black/(n_total-2)) + (n_black/(n_total-1))*(n_red/(n_total-2)))

print(p_each_color_once_v1, p_each_color_once_v2, p_each_color_once_v3)

# could have done this more easily, doh
# this probably works because we're using all available colours anyway (and including the permutations in each branch),
# so the order of choice balances itself out
p_3_unique = 3 * p_each_color_once_v1

print(f'The probability of picking a different colour each time with three picks is {p_3_unique}.')
```

1. $p_{\text{white}} = \frac{n_{\text{white}}}{n_{\text{total}}} = 0.\bar{3}$ 
2. $p_{\text{white,black,dep}} = p_{\text{white}} * \frac{n_\text{black}}{n_\text{total} - 1}$
3. $p_{\text{white,black,indep}} = p_{\text{white}} * \frac{n_\text{black}}{n_\text{total}}$
4. $p_{\text{3 distinct colours}} = 3 * p_{\text{colour1}} (\frac{n_\text{colour2}}{n_\text{total}-1} * \frac{n_\text{colour3}}{n_\text{total}-2} + \frac{n_\text{colour3}}{n_\text{total}-1} * \frac{n_\text{colour2}}{n_\text{total}-2}) = 0.236$

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
p_cloudy = .4
p_rainy_and_cloudy = .5
p_rainy_and_ncloudy = .2

p_rainy = p_cloudy * p_rainy_and_cloudy + (1 - p_cloudy) * p_rainy_and_ncloudy

print(f'The likelihood of rain next Saturday on the given parameters is at {p_rainy * 100}%.')
```

$P(rainy) = P(cloudy) * P(rainy|cloudy) + P(\lnot cloudy) * P(rainy|\lnot cloudy) = 0.32$

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# YOUR CODE HERE
p_a = 0.3
p_b = 0.6
p_a_AND_b = 0.1

p_not_a = 1-p_a
print('p_not_a', p_not_a)
p_a_OR_b = p_a + p_b - p_a_AND_b
print('p_a_OR_b', p_a_OR_b) # that's an odd rounding error
p_a_AND_NOT_b = p_a - p_a_AND_b
print('p_a_AND_NOT_b', p_a_AND_NOT_b)
p_a_OR_NOT_a = 1
print('p_a_OR_NOT_a', p_a_OR_NOT_a)
p_a_OR_b_AND_NOT_b = p_a
print('p_a_OR_b_AND_NOT_b', p_a_OR_b_AND_NOT_b)
p_NOT_a_AND_NOT_b = 1- (p_a + p_b - p_a_AND_b)
print('p_NOT_a_AND_NOT_b', p_NOT_a_AND_NOT_b)
p_NOT_a_AND_b = 1-p_a_AND_b
print('p_NOT_a_AND_b', p_NOT_a_AND_b)

```

* $P(\overline{A}) = 0.7$  $\Rightarrow$ **True**
* $P(A \cup B) = 0.6$ $\Rightarrow$ **False**, should be $P(A)+P(B)-P(A \cap B) = 0.3 + 0.6 - 0.1 = 0.8$
* $P(A \cap \overline{B}) = 0.2$ \Rightarrow$ **True**
* $P(A \cup \overline{A}) = 1$  $\Rightarrow$ **True** (Tautology?)
* $P(A \cap (B \cup \overline{B})) = 0.4$ \Rightarrow$ **False**, the intersection of A and everything is A (which is 0.3)
* $P(\overline{A} \cap \overline{B}) = 0.3$ \Rightarrow$ **True**
* $P(\overline{A \cap B}) = 0.9$ \Rightarrow$ **True**

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE
p_covid_china = 0.01
p_covid_italy =0.2

p_fly_italy = 1/6
p_fly_china = 5/6


p_covid = p_covid_china * p_fly_china + p_covid_italy * p_fly_italy

p_china_covid = p_covid_china * p_fly_china / p_covid

print(f'The probability that you flew to China given that you contracted Covid is {p_china_covid}.')
```

$P(China|Covid) = \frac{P(Covid|China) * P(China)}{P(Covid)} = \frac{P(Covid|China) * P(China)}{P(Covid|China) * P(China) + P(Covid|Italy) * P(Italy)}= 0.2$
